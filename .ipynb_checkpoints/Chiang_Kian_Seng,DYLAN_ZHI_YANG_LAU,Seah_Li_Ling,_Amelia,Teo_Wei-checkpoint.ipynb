## Project Statement
In the last 40 years, Singapore has been facing a gradual declining fertility rate; from 1.82 in the 1980 to 1.14 in 2019. This trend is becoming increasingly worrying for the Singapore government, with numerous local and international media coverage on this phenomenon. 
Just like how in every social issue, there is no one reason that could be fully attributed as the cause of the problem. Hence, our group is curious about the underlying dimensions, even seemingly unrelated ones, that is causing the decline in Singapore’s annual birth rate. 

## Project Objective
Our project aims to investigate the relationship between Singapore’s birth rate from 2009 - 2019, and a few social and economic dimensions. The social and enconomic dimensions we will be investigating are:
1. Females’ education levels.
2. Females’ median age of first marriages.
3. Cost of some consumer goods and services. 
4. Average median monthly household income.

that could potentially affect Singapore's annual birth rate to an extent.

## Hypothesis
Our hypothesis is that following factors could be contributing to lower birth rates in Singapore: 
1. __Social Factors__ - the rising median age of females first marriages and the rising proportion of females with higher education qualifications
2. __Economic Factors__ - rising cost of some consumer items like Formula Milk Powder, Accommodations, Specialist Outpatient Clinic Fees, Hospital Services and Education

```python
df_female_highest_edu.reset_index(inplace=True, drop=True)
df_female_highest_edu.head()
```

3. As we are interested only in the female population, we will extract the following columns
    1. Total Female Count
    2. Below Secondary Education count
    3. Secondary Education count
    4. Post Secondary (Non-Tertiary) Education Count
    5. Diploma & Professional Qualification count
    6. University Education count

```python
# Create a temporary dataframe to store the extracted columns
df_temp = pd.DataFrame(df_female_highest_edu['Subject: Education, Language Spoken and Literacy'])
# df_temp
```

```python
# Function to loop through all the columns, searching row index 0 for data labels that we require.
# The function will then extract the column when found and concat to the temporary dataframe
def extract_columns(in_df, out_df):
    # List of row index 0 values we want
    value_names = ['Total Females', 'Below Secondary', 
                   'Secondary', 'Post Secondary (Non-tertiary)',
                   'Diploma & Professional Qualification', 'University'
                  ]
    
    for value in value_names:
        for col in in_df.columns:
            if in_df[col].isin([value]).any():
                # When a match is found, extract the column and concat to our temporary dataframe
                out_df = pd.concat([out_df, pd.DataFrame(in_df[col])], axis=1, copy=True)
                break
            else:
                # Drop columns when we have not found what we want
                in_df.drop(columns=col, inplace=True)

    return out_df
```

```python
# Call function to extract the required columns of data.
df_temp = extract_columns(df_female_highest_edu, df_temp)
# display(df_female_highest_edu.head())
display(df_temp.head())
```

4. Change column names to appropriate ones and drop row index 0 as it is no longer required

```python
# Change column names to appropriate names
new_col_map = {
    'Subject: Education, Language Spoken and Literacy': 'Year',
    'Unnamed: 121': 'Total Females',
    'Unnamed: 131': 'Below Secondary',
    'Unnamed: 141': 'Secondary',
    'Unnamed: 151': 'Post Secondary',
    'Unnamed: 161': 'Diploma and Professional Qualification',
    'Unnamed: 171': 'University'
}

df_temp.rename(columns=new_col_map, inplace=True)
display(df_temp.head())
```

```python
# Drop row index 0
df_temp.drop(index=0, inplace=True)
display(df_temp.head())
```

5. Check for null values and remove rows that have them

```python
# Remove all rows with null values
display(df_temp.isnull().sum())
df_temp.dropna(inplace=True)
display(df_temp.isnull().sum())
df_temp
```

6. Check and convert data types to appropriate ones

```python
# Check and convert data types
display(df_temp.dtypes)
for col in df_temp.columns:
    df_temp[col] = df_temp[col].astype('int64')
print('After converting...')
display(df_temp.dtypes)
```

7. Finally, overwrite our origianl dataframe with temporary one that has all the correct data

```python
df_female_highest_edu = df_temp
df_female_highest_edu.head()
```

3. We are interested in the CPIs of the items with the following sub-level column names:
    1. Formula Milk Powder
    2. Accomodation
    3. Fees At Specialist Outpatient Clinics
    4. Hospital Services
    5. Education<br><br>
   The code below will loop through each column in the CPI dataframe looking for the above sub-level column names, extract the column data and concat to a temporary dataframe

```python
in_df  = df_cpi.copy()
out_df = pd.DataFrame(df_cpi['Subject: Prices and Price Indices'])
# display(in_df.head())
# display(out_df.head())

values_we_want = ['Formula Milk Powder', 'Accommodation', 'Fees At Specialist Outpatient Clinics',
                 'Hospital Services', 'Education']
for value in values_we_want:
    for col in in_df.columns:
        if in_df[col].isin([value]).any():
            out_df = pd.concat([out_df, pd.DataFrame(in_df[col])], axis=1)
            break
out_df.reset_index(inplace=True, drop=True)
display(out_df)
```

4. Remove row index 0 to 3 as they are no longer needed

```python
out_df.drop(np.arange(4), inplace=True)
out_df.head()
```

5. Rename the columns to the appropriate names

```python
col_index = 0
col_map = {}
for col in out_df.columns:
    if col_index == 0:
        col_map[col] = 'Year'
    else:
        col_map[col] = values_we_want[col_index - 1]
    col_index += 1
    
# display(col_map)
out_df.rename(columns=col_map, inplace=True)
display(out_df.head())
```

6. Remove all rows with null values

```python
display(out_df.isnull().sum())
out_df.dropna(inplace=True)
display(out_df.isnull().sum())
display(out_df)
```

7. Extract rows from year 2014 onwards as they have usable data for all columns

```python
# First convert 'Year' column to int64
out_df['Year'] = out_df['Year'].astype('int64')
# Extract rows from year 2014 onwards
out_df = out_df[out_df['Year'] >= 2014]
display(out_df)
```

8. Convert to appropriate data types

```python
print("Before conversion...")
display(out_df.dtypes)
out_df['Formula Milk Powder'] = out_df['Formula Milk Powder'].astype('float64')
out_df[values_we_want] = out_df[values_we_want].astype('float64')
print('Afer Conversion...')
display(out_df.dtypes)
```

9. Overwrite original dataframe

```python
df_cpi = out_df
display(df_cpi)
```

### Fertility Data Visualization

```python
# recap what the fertility dataframe looks like
df_fertility.head()
```

```python
fig = plt.figure(figsize=(16,16))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)
df_fertility.plot(kind='scatter', x='Year', y='Total Fertility Rate *(Per Female)', 
                  marker='D', color='tomato', linewidth=2, ax=ax1)
x = df_fertility['Year']
y = df_fertility['Total Fertility Rate *(Per Female)']
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
ax1.plot(x, p(x), linestyle=':', color='darkorchid', linewidth=4)
ax1.legend(['Trend Line', 'Total Fertility Rate *(Per Female)'])
ax1.set_title('Graph 1: Trend of Annual Total Fertility Rates', fontsize=16)
ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Total Fertility Rate *(Per Female)', fontsize=14)

df_fertility.plot(kind='line', x='Year', 
                  y=['Age 15-19', 'Age 20-24', 'Age 25-29', 'Age 30-34', 
                     'Age 35-39', 'Age 40-44', 'Age 45-49'], 
                  ax=ax2, marker='D', linewidth=2)
ax2.set_title('Graph 2: Trend of Fertility Rates by Age Group', fontsize=16)
ax2.set_xlabel('Year', fontsize=14)
ax2.set_ylabel('Age-specific Fertility Rate (Per 1000 females)', fontsize=14)

plt.show()
```

#### Observations & Insights From the Above Graphs

3. From Graph2, we also observe that the fertility rate per 1000 females is climbing for the age groups 30 to 34 and 35 to 39

### Other observations
We also want to compare the changes in the proportion of the female population attaining each of the education qualification for 2009 and 2019 (a decade apart)

```python
# check the proportions values for 2019 vs 2009 (a decade ago)
df_decade = df_female_highest_edu.set_index('Year', drop=True)
df_decade = df_decade.loc[[2009, 2019]][['Below Secondary Pct', 'Secondary Pct', 'Post Secondary Pct',
                                        'Diploma and Professional Qualification Pct',
                                        'University Pct']]
print('Extracting data from years 2009 and 2019:')
display(df_decade)
pct_changes = pd.DataFrame(((df_decade.loc[2019] / df_decade.loc[2009]) - 1) * 100, columns=['Percent Change (%)'])
print("Percent change in proportions in 2019 compared to 2009:")
display(pct_changes)
```

**From the tables above, we can see that there is a significant increase in the proportion of the female population attaining University qulifications as compared to 2009. In 2009, it was *20.3%* of the female population and it increased to *30.8%* in 2019. This is an increase of a whopping *51.7%*. Next biggest increases are Post Secondary and Diploma and Professional Qualifications with increases of *44.4%* and *30.0%*, respectively.**

**The above indicates that a larger proportion of the female population are seeking to have higher education qualifications in 2019 as compared to 2009. This would mean that a larger proportion of the female population are incurring longer schooling years in 2019 as compared to 2009, leading to larger proportion of the female population delaying their marriages and thus their child bearing years to a later age. This could translate to lower Total Fertility Rate as the success rate of child bearing decreases as women age.**

### Correlation Study Of Various Social Factors and Total Fertility Rate 
We will merge the following data from the various datasets to form a correlation study dataset:
1. Total Fertility Rate from the Fertility dataset
2. Overall brides first marriage median age
3. Proportion of females highest attained education qualification from the Female Highest Attained Education Qualification.
4. Household Incomes of at least one employed member from the Household incomes dataset

<b>Note: As the Household incomes dataset starts in the year 2000, after merging these dataset, our merged dataset will also begin in the year 2000</b>

#### Preparing the correlation dataset

```python
# Extract only the Year and Total Fertility Rate columns
df_tfr = df_fertility[['Year', 'Total Fertility Rate *(Per Female)']]
# display(df_tfr.head())
df_mma = df_marriage_median_age[['Year', 'overall_brides_first_marriage_median_age']]
# display(df_mma.head())
df_uni_edu = df_female_highest_edu[['Year', 'Below Secondary Pct', 'Secondary Pct', 'Post Secondary Pct', 'Diploma and Professional Qualification Pct','University Pct']]
# display(df_uni_edu.head())
df_emp_inc = df_household_income[['Year', 'employed_household_median_income']]
# display(df_emp_inc.head())
df_corr = pd.merge(df_tfr, df_mma)
df_corr = pd.merge(df_corr, df_uni_edu)
df_corr = pd.merge(df_corr, df_emp_inc)
display(df_corr.head())
```

#### Displaying the Social Factors correlation table

```python
# Display the correlation table
df_corr_tbl = df_corr.copy()
df_corr_tbl.drop(columns=['Year'], inplace=True)
# display(df_corr_tbl.head())
df_corr_tbl = df_corr_tbl.corr()
# display(df_corr_tbl)
df_corr_tbl = df_corr_tbl.style.background_gradient(cmap='Reds')
df_corr_tbl = df_corr_tbl.set_precision(2)
df_corr_tbl
```

#### Observations and Insights From the Correlation Study of Various Social Factors and Total Fertility Rate
1. The decline is **Total Fertility Rate** has a relatively high negative correlation with **Brides First Marriage Median Age** at -0.78.<br>**This seems to support our premise that as the trend of median age of female first marriages goes up, the Total Fertility Rate tends to trend downwards.**
<br>
2. The decline in **Total Fertility Rate** also has a relatively high negative correlation with the **proportion of females with the following higest attained qualifications**:
    1. University Pct at -0.79
    2. Diploma and Professional Qualification Pct at -0.73<br>
    
    <b>This seems to supports our premise that as the proportion of the female population go for higher education qualifications trends upwards, the Total Fertility Rate trends downwards</b>
<br>    
3. There is a strong positive correlation of the rising trend of the **Median Age of Females First Marriages** with the rising trend of the **proportions of females going for higher education qualifications**. The 2 categories of higher education qualification are:
    1. Diploma and Professional Qualification Pct at 0.99
    2. University Pct at 0.99<br>
    
    <b>This seems to support our premise that as the trend of proportions of the female population seeking higher education qualifications, they tend to also delay their marriages to a later age, in turn affecting the Total Fertility Rate</b>
<br>    
4. There is strong positive correlation between the rising trend of the **Median Age of Females First Marriages** and the rising trend of **Household Income with at least One Employed Member** at 0.98.<br> 
    **This seems to support our premise that women tend to increasingly stay employed in their early child bearing ages before deciding to stop working to have a family at a later age**
